```python
afstemning = afstemning.drop(["nummer", "konklusion", "kommentar", "opdateringsdato"], axis=1)
```

```python
afstemning = afstemning.drop(["møde_status", "møde_titel", "møde_type", "sagstrin_sag_titel", "sagstrin_sag_titelkort"], axis=1)
```

```python
afstemning = afstemning.drop(["møde_periode_type", "type", "sagstrinid"], axis=1)
```

```python
afstemning = afstemning.drop(["sagstrin_titel", "sagstrin_status", "sagstrin_type", "sagstrin_sag_type"], axis=1)
```

```python
afstemning = afstemning.drop(["mødeid", "typeid", "sagstrin_sag_kategori", "sagstrin_sag_status", "sagstrin_periode_type"], axis=1)
```

```python
afstemning = afstemning[~afstemning["text"].isna()]
```

```python
afstemning["tokens"] = afstemning.loc[:,"text"].apply(lambda x: x.split())
```

```python
import spacy
from spacy.lang.da.examples import sentences 

nlp = spacy.load("da_core_news_sm")
from collections import Counter

from nltk.corpus import stopwords
DA_STOP_WORDS = stopwords.words("danish")
from spacy.lang.da.stop_words import STOP_WORDS
STOP_WORDS = list(STOP_WORDS) + DA_STOP_WORDS
import string 
STOP_WORDS = list(string.punctuation) + STOP_WORDS
```

```python
afstemning["text"] = afstemning["text"].apply(lambda x: x.strip())
tokens = list(map(nlp, afstemning["text"].values))
```

```python
words = [[token.text for token in doc if token.text not in STOP_WORDS] for doc in tokens]
```

```python
cnt = Counter()
for alist in words:
    cnt.update(alist)
```

```python
sorted(cnt.items(), key=lambda k: k[1], reverse=True)
```
